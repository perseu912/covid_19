```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker


#importando os dados do covid 19 em juazeiro
dados_jua = pd.read_csv('jua.csv',sep =',')
print(dados_jua)
dados_jua.to_excel('jua.xlsx')
```

```python
dados_jua = dados_jua.loc[::-1]
dados_jua.plot(x='date',y='new_confirmed')
```

```python
dados_pet = dados_pet.loc[::-1]
dados_pet.plot(x='date',y='new_confirmed')
```

```python

def calcSma(data, smaPeriod):
    j = next(i for i, x in enumerate(data) if x is not None)
    our_range = range(len(data))[j + smaPeriod - 1:]
    empty_list = [None] * (j + smaPeriod - 1)
    sub_result = [np.mean(data[i - smaPeriod + 1: i + 1]) for i in our_range]

    return np.array(empty_list + sub_result)
```

```python
meses = ['janeiro','fevereiro','maro','abril','maio','junho','julho','agosto','setembro','outubro','novembro','dezembro']
m_c = []
for dat in dados_jua['date']:
    dt = dat[5:7]
    #print(dt)
    m = meses[1*int(dt)-1]
    #print(m)
    m_c.append(m)
dados_jua['meses'] = m_c
dados_pet['meses'] = m_c
dados_jua.plot(x='meses',y='new_confirmed')
```

```python
casos_diarios = dados_jua['new_confirmed']
media = calcSma(casos_diarios,7)

ax = dados_jua.plot(x='meses',y='new_confirmed',label='casos dirios',kind='bar',color='black')


plt.plot(dados_jua['date'],media,label='media mvel',c='red')
# Make most of the ticklabelhs empty so the labels don't get too crowded
ticklabels = ['']*len(dados_jua.index)
# Every 4th ticklable shows the month and day
ticklabels[::30] = dados_jua['meses'][::30]
# Every 12th ticklabel includes the year
ax.xaxis.set_major_formatter(ticker.FixedFormatter(ticklabels))
#plt.grid()
plt.gcf().autofmt_xdate()
print('covid-19 em Juazeiro-Ba')
plt.ylabel('casos confirmados')
plt.title('covid-19 em Juazeiro')
plt.grid(axis='y')
plt.legend()
plt.savefig('plot_media_movel_jua.png',dpi=900)
plt.show()
```

```python
casos_diarios = dados_pet['new_confirmed']
media = calcSma(casos_diarios,7)

ax = dados_pet.plot(x='meses',y='new_confirmed',label='casos diarios',kind='bar',color='black')

#plt.grid()
plt.plot(dados_pet['date'],media,label='media mvel',c='red')
# Make most of the ticklabelhs empty so the labels don't get too crowded
ticklabels = ['']*len(dados_pet.index)
# Every 4th ticklable shows the month and day
ticklabels[::30] = dados_pet['meses'][::30]
# Every 12th ticklabel includes the year
ax.xaxis.set_major_formatter(ticker.FixedFormatter(ticklabels))
plt.gcf().autofmt_xdate()
#plt.grid()
print('covid-19 em Petrolina-PE')
plt.title('covid-19 em Petrolina')
plt.ylabel('casos confirmados')
plt.legend()
plt.grid(axis='y')
plt.savefig('plot_media_movel_pet.png',dpi=800)
plt.show()
```

```python
mortes_diarios = abs(np.array(dados_pet['new_deaths']))
media = calcSma(mortes_diarios,7)

ax = dados_pet.plot(x='meses',y='new_deaths',label='mortes dirias',kind='bar',color='black')
ax.set_ylim(0,7)
#plt.grid()
plt.plot(dados_pet['date'],media,label='media mvel',c='red')
# Make most of the ticklabelhs empty so the labels don't get too crowded
ticklabels = ['']*len(dados_pet.index)
# Every 4th ticklable shows the month and day
ticklabels[::30] = dados_pet['meses'][::30]
# Every 12th ticklabel includes the year
ax.xaxis.set_major_formatter(ticker.FixedFormatter(ticklabels))
plt.gcf().autofmt_xdate()
#plt.grid()
print('covid-19 em Petrolina-PE')
plt.title('covid-19 em Petrolina')
plt.ylabel('mortes confirmadas')
plt.legend()
plt.grid(axis='y')
plt.savefig('plot_morte_media_movel_pet.png',dpi=800)
plt.show()
```
