```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker


#importando os dados do covid 19 em juazeiro
dados_jua = pd.read_csv('jua.csv',sep =',')
print(dados_jua)
dados_jua.to_excel('jua.xlsx')
```

```python
dados_jua = dados_jua.loc[::-1]
dados_jua.plot(x='date',y='new_confirmed')
```

```python

def calcSma(data, smaPeriod):
    j = next(i for i, x in enumerate(data) if x is not None)
    our_range = range(len(data))[j + smaPeriod - 1:]
    empty_list = [None] * (j + smaPeriod - 1)
    sub_result = [np.mean(data[i - smaPeriod + 1: i + 1]) for i in our_range]

    return np.array(empty_list + sub_result)
```

```python
casos_diarios = dados_jua['new_confirmed']
media = calcSma(casos_diarios,7)

ax = dados_jua.plot(x='date',y='new_confirmed',kind='bar')

plt.plot(dados_jua['date'],media,label='media mvel',c='black')
# Make most of the ticklabelhs empty so the labels don't get too crowded
ticklabels = ['']*len(dados_jua.index)
# Every 4th ticklable shows the month and day
ticklabels[::30] = dados_jua['date'][::30]
# Every 12th ticklabel includes the year
ax.xaxis.set_major_formatter(ticker.FixedFormatter(ticklabels))
plt.gcf().autofmt_xdate()
print('covid-19 em Juazeiro-Ba')
plt.title('covid-19 em Juazeiro')
plt.savefig('plot_media_movel_jua.png',dpi=800)
plt.show()
```

```python
import numpy as np
confirmados = dados_jua['last_available_confirmed']
#print(np.array(confirmados))
percent = (np.array(confirmados)[-1] - np.array(confirmados)[-30])/np.array(confirmados)[-30]

print(round(percent*100,2),'%')
media = calcSma(casos_diarios[-30:],6)
plt.bar(np.linspace(1,30,30),casos_diarios[-30:],label='casos diarios em Novembro')
plt.plot(np.linspace(1,30,30),media,label='media movel',c='black')
plt.xlabel('dia')
plt.ylabel('confirmados')
plt.legend()
plt.show()
import numpy as np
plt.bar(np.linspace(1,30,30),casos_diarios[-60:-30],label='casos diarios em Outubro')
media = calcSma(casos_diarios[-30:],6)
plt.plot(np.linspace(1,30,30),media,label='media movel',c='black')
plt.xlabel('dia')
plt.ylabel('confirmados')
plt.legend()
plt.show()
```
