```python
city = 'juazeiro'

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker


dados_pet = pd.read_csv(f'dados_{city}.csv',sep=',')
#dados_pet.plot()

confirmados = dados_pet['last_available_confirmed'][::-1]
mortos = dados_pet['last_available_deaths'][::-1]
casos_diarios = dados_pet['new_confirmed'][::-1]



meses = ['jan','fev','mar','abr','mai','jun','jul','ago','set','out','nov','dez']
date = dados_pet['date'][::-1]
date = [i[4:] for i in date]
print(int(date[0][2:3]))
date = [('/').join([i[4:],meses[int(i[1:3])-1]]) for i in date]
date




dados = {'confirmados':np.array(confirmados),
         'mortos':np.array(mortos),
         'casos_diarios':np.array(casos_diarios),
         'date':date}
dados = pd.DataFrame(dados,index=date)
dados.to_excel(f'dados_{city}.xlsx')
dados





#dados[-30:].plot(y=['casos_diarios'],kind='bar')




def calcSma(data, smaPeriod):
    j = next(i for i, x in enumerate(data) if x is not None)
    our_range = range(len(data))[j + smaPeriod - 1:]
    empty_list = [None] * (j + smaPeriod - 1)
    sub_result = [np.mean(data[i - smaPeriod + 1: i + 1]) for i in our_range]

    return np.array(empty_list + sub_result)


ax = dados.plot(x='date',y='casos_diarios',kind='bar',c='black')

media = calcSma(casos_diarios,7)
plt.plot(dados['date'],media,label='media m√≥vel',c='red')

# Make most of the ticklabels empty so the labels don't get too crowded
ticklabels = ['|']*len(dados.index)
# Every 4th ticklable shows the month and day
ticklabels[::20] = dados['date'][::20]
# Every 12th ticklabel includes theear
ax.xaxis.set_major_formatter(ticker.FixedFormatter(ticklabels))
plt.gcf().autofmt_xdate()
print(f'covid-19 em {city}')
plt.title(f'covid-19 em {city}')
plt.savefig(f'plot_media_{city}.png',dpi=800)
plt.show()

```
